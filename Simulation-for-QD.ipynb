# 1. Prepare

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from math import comb
from IPython.display import Image
from numpy.random import randint, default_rng
%config InlineBackend.figure_format='retina'
```

```python
def gen_list(S, N, T):
    result_list = []
    for D in range(1, N + 1):
        result = 0
        for k in range(T + 1, S + 1):
            A = comb(N - D, k)

            B = 0
            for i in range(k, S + 1):
                for j in range(k, S + 1):
                    if i + j - k <= N - D:
                        B1 = comb(D, S - i)
                        B2 = comb(N - D - k, i - k)
                        B3 = comb(N - D - i, j - k)
                        B4 = comb(D, S - j)
                        B += B1 * B2 * B3 * B4

            C = comb(N, S) * comb(N, S)
            result += A * B / C
        result_list.append(result)
    return result_list
```

# 2. Small Sample Result

```python
y1 = gen_list(S=70, N=100, T=10)
y2 = gen_list(S=70, N=100, T=20)
y3 = gen_list(S=70, N=100, T=30)
y4 = gen_list(S=70, N=100, T=40)
x = list(range(1, 100 + 1))
```

```python
plt.plot(x, y1, label='T=10')
plt.plot(x, y2, label='T=20')
plt.plot(x, y3, label='T=30')
plt.plot(x, y4, label='T=40')
plt.legend()
plt.xlabel('D')
plt.ylabel('Q(D)')
plt.title('Q(D) Curve when S=70, N=100')
```

# 3. Large Sample Result
